```python
cat= [18,40,80,100,120]
lab=[1,2,3,4]
data["offer"]= pd.cut(data['MonthlyCharges'], bins=cat , labels=lab)
```

```python
data.head(4)
```

```python
def offer2(row):
    if row['gender']=="Male" and row['SeniorCitizen']==1:
        return "Male_senior"
    elif row['gender']=="Male" and row['SeniorCitizen']==0:
        return "Male_junior"
    elif row['gender']=="Female" and row['SeniorCitizen']==1:
        return "Female_senior"
    elif row['gender']=="Female" and row['SeniorCitizen']==0:
        return "Female_junior"
    else:
        return "other"
        
```

```python
data['offer2']= data.apply(offer2, axis = 1)
```
