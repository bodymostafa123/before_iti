```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('drive/My Drive/EURUSD60.csv',sep='\t',names=['Date','open','high','low','Close','Volume'])
```

```python
data.head()
```

```python
f ,ax =plt.subplots(figsize=(11,8))
data['Close'].plot(grid=True)
```

```python
data['moveAv50']=data.loc[:,'Close'].rolling(window=50).mean()
data['moveAv70']=data.loc[:,'Close'].rolling(window=70).mean()

data.head()
```

```python
data.head(70)
```

```python
data=data.iloc[70:,:]
data.head()
```

```python
data['AdjustClose']=data['Close'].shift(-1)
data.head()
```

```python
data=data.loc[:,['Date','moveAv50','moveAv60','AdjustClose']]
data.head()
```

```python
data.tail()
```

```python
data=data.iloc[:-1,:]
data.head()
```

```python
data.tail()
```

```python
data.set_index(['Date'],inplace=True)
data.head()
```

```python
n_step=1
n_features=2
values=data.values
n_train=int(len(data)*0.8)
train=values[:n_train,:]
test=values[n_train:,:]

train_X,train_y=train[:,:-1],train[:,-1]

test_X,test_y=test[:,:-1],test[:,-1]

```

```python
train_X.shape
```

```python
train_X=train_X.reshape(train_X.shape[0],n_step,n_features)
test_X=test_X.reshape(test_X.shape[0],n_step,n_features)
train_X.shape
```

```python
train_y.shape
```

```python
train_y=train_y.reshape(-1,1)
test_y=test_y.reshape(-1,1)
train_y.shape
```

Build RNN model

```python
from tensorflow import keras 
from keras import Sequential

from keras.layers import LSTM,Dense

model=Sequential()

model.add(LSTM(150,activation='relu',input_shape=(train_X.shape[1],train_X.shape[2])))
model.add(Dense(50,activation='relu'))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse')
historty=model.fit(train_X,train_y,batch_size=64,epochs=20,validation_split=0.15)
```

```python
plt.plot(historty.history['loss'],label='train')
plt.plot(historty.history['val_loss'],label='validation')
plt.legend()
plt.show()
```

```python
y_pred=model.predict(test_X)

from sklearn.metrics import mean_squared_error

print("MSE : ",mean_squared_error(y_pred,test_y))
```

```python
plt.figure(figsize=(18,6))
plt.plot([x for x in range(1,y_pred.shape[0]+1)],y_pred,linestyle='-',color='red',label='prediction')
plt.plot([x for x in range(1,test_y.shape[0]+1)],test_y,linestyle='-',color='blue',label='actual')
plt.legend()
plt.show()
```
