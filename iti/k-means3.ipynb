#Question 3 Apply elbow method

```python
L=[]
for i in range(2,10):
    img2=img_flat.copy()
    model=KMeans(n_clusters=i)
    model.fit(img_flat)
    for j in np.unique(model.labels_):
        img2[model.labels_==j,:]=model.cluster_centers_[j]
    L.append(model.inertia_)
    img2=img2.reshape(img.shape)
    plt.imshow(img2) 
    plt.axis('off')
    plt.title('clusters: '+str(i))
    plt.show()
```

```python
plt.plot(L)
```
